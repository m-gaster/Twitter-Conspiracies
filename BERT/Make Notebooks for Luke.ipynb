```python
def gen_split_n_notebook(new_or_old, name, split, new_read_path):
    
    save_ipynb_path = NEW_NOTEBOOKS_PATH_ROOT + f'{name} - {new_or_old} Split {split}.ipynb'
    print(save_ipynb_path)
    
    with open(NOTEBOOK_TEMPLATE_PATH, 'r') as fp:
        notebook = json.load(fp)
    
    for cell in notebook['cells']:
        if cell['cell_type']=='code':
            cell['source'] = [x.replace('REPLACE_READ_PATH_HERE', new_read_path) for x in cell['source']]
            
    with open(save_ipynb_path, 'w') as fp:
        json.dump(notebook, fp)
```

```python
for split in range(0,20):
    for name in ('Luke', 'Mikhael'):
        
        new_path = fr"/arc/project/st-tlemieux-1/data/Twint-Data/{name}/New-Splits/Split {split}/"

        gen_split_n_notebook(new_or_old='New',
                             name=name,
                             split=split,
                             new_read_path='r"' + new_path + '"')
        
        
for split in range(0,5):
    
    new_path = fr"/arc/project/st-tlemieux-1/data/Twint-Data/Luke/Split {split}/"
    
    gen_split_n_notebook(new_or_old='Old',
                         name='Luke',
                        split=split,
                        new_read_path='r"' + new_path + '"')
    
for split in range(5,10):
    
    new_path = fr"/arc/project/st-tlemieux-1/data/Twint-Data/Mikhael/Split {split}/"
    
    gen_split_n_notebook(new_or_old='Old',
                         name='Mikhael',
                        split=split,
                        new_read_path='r"' + new_path + '"')
```
