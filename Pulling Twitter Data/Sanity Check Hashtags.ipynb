```python
import pandas as pd
import ast
```

```python
file_path = "C:/Users/lukea/Documents/GitHub/twint/data/4/Split-4.csv1124368051028275200_TWEETS.csv"

df = pd.read_csv(file_path)
hashtag_list = [
    'plandemic',
    'scamdemic',
    'covidhoax',
    'nwo',
    'covid1984',
    'plandemia',
    'agenda21',
    'thegreatreset',
    'agenda2030',
    'newworldorder',
    'wakeupamerica',
    'wakeup',
    'openamericanow',
    'firefauci',
    'wwg1wga',
    'qanon',
    'coronahoax'
]
```

Check list of hashtags used:

```python
hashtags_used = df['hashtags'].apply(ast.literal_eval).explode().value_counts().to_dict()
hashtags_used
```

Are any in the list of conspiracy hashtags?

```python
count = 0
for each in hashtags_used:
    if each in hashtag_list:
        count += 1   
count
```

```python
count = 0
for tweet in df['tweet']:
    if any(word in tweet for word in hashtag_list):
        count += 1
count
```
