```python
def recover_searched_ids(path):
    
    searched_users = [int(f.split('_')[0]) for f in listdir(path) if isfile(join(path, f))]
     
    return searched_users
```

```python
def recreate_search_updated_dataset(split):
    '''
    split = int detailing which split of the dataset we're recovering
    '''   
    
    split_n_path = r"C:/Users/crackcocaine69xxx/Python Stuff/594/Twint/Looking Up All Conspiracy Hashtag User Tweets/All Conspiracy Tweeters' Tweets/Split" + " " + str(split)
    
    searched_users = np.array(recover_searched_ids(split_n_path))
    
    split_n = pd.read_csv(r"C:/Users/crackcocaine69xxx/Python Stuff/594/Twint/Getting Conspiracy Hashtag Users/All User IDs.csv").set_index('ID')
    
    split_n = split_n.iloc[split::10]
    
    split_n['Searched'].loc[searched_users] = 1

    
    overwrite_path = r"C:/Users/crackcocaine69xxx/Python Stuff/594/Twint/Getting Conspiracy Hashtag Users/All User IDs - Split" + " " + str(split) + ".csv"
    
    split_n = split_n.reset_index()
    
    split_n['Error'] = 0
    
    split_n.to_csv(overwrite_path)
```
